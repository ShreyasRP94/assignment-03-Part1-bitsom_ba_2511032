# **#Task 1**

```python
# Raw Data provided in the problem statement

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

print(raw_students)
```

```python
# Name cleaning separately

for s in raw_students:
    original = s["name"]
    cleaned = original.strip().title()
    print(f"Original: '{original}' -> Cleaned: '{cleaned}'")
```

```python
# Marks conversion

for s in raw_students:
    raw_marks = s["marks_str"]
    split_marks = raw_marks.split(",")
    
    marks = []
    for m in split_marks:
        marks.append(int(m))
    
    print(raw_marks, "->", marks)
```

```python
# Building cleaned students dataset

students_clean = []

for s in raw_students:
    
    # Name
    name_clean = s["name"].strip().title()
    
    # Roll
    roll_num = int(s["roll"])
    
    # Marks
    marks_temp = s["marks_str"].split(",")
    marks = []
    for m in marks_temp:
        marks.append(int(m))
    
    student_obj = {
        "name": name_clean,
        "roll": roll_num,
        "marks": marks
    }
    
    students_clean.append(student_obj)

students_clean
```

```python
# Name validity test

for s in students_clean:
    name = s["name"]
    
    valid = True
    words = name.split()
    
    for w in words:
        if not w.isalpha():
            valid = False
    
    if valid:
        print(name, "- ✓ Valid name")
    else:
        print(name, "- ✗ Invalid name")
```

```python
# Profile card display

for s in students_clean:
    
    print("=" * 32)
    print(f"Student : {s['name']}")
    print(f"Roll No : {s['roll']}")
    print(f"Marks   : {s['marks']}")
    print("=" * 32)
```

```python
# Print the name in ALL CAPS and lowercase for the student with roll number 103

for s in students_clean:
    if s["roll"] == 103:
        name = s["name"]
        print("Uppercase :", name.upper())
        print("Lowercase :", name.lower())
```

# **#Task 2**

```python
# Data provided in the Task 2

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
# Subject-wise grades 

print("Subject-wise Performance:\n")

for i in range(len(subjects)):
    sub = subjects[i]
    score = marks[i]
    
    # Grade logic
    if 90 <= score <= 100:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{sub:10} : {score} -> {grade}")
```

```python
#Printing important metrics

# Total
total_marks = sum(marks)

# Average
avg_marks = round(total_marks / len(marks), 2)

# Highest
max_marks = max(marks)
max_index = marks.index(max_marks)
top_subject = subjects[max_index]

# Lowest
min_marks = min(marks)
min_index = marks.index(min_marks)
low_subject = subjects[min_index]

print("\nSummary:")
print("Total Marks       :", total_marks)
print("Average Marks     :", avg_marks)
print("Highest Subject   :", top_subject, "-", max_marks)
print("Lowest Subject    :", low_subject, "-", min_marks)
```

```python
# While loop

new_count = 0

while True:
    print("\n--- Marks Entry Menu ---")
    print("1. Add new subject")
    print("2. Stop entry")
    
    choice = input("Enter choice (1/2): ").strip()
    
    if choice == "2":
        break
    
    elif choice == "1":
        subject_input = input("Enter subject name: ").strip()
        
        if subject_input == "":
            print("Subject name cannot be empty")
            continue
        
        marks_input = input(f"Enter marks for {subject_input}: ").strip()
        
        # Validation
        if not marks_input.isdigit():
            print("Marks must be numeric")
            continue
        
        marks_value = int(marks_input)
        
        if marks_value < 0 or marks_value > 100:
            print("Marks must be between 0 and 100")
            continue
        
        # Add valid entry
        subjects.append(subject_input)
        marks.append(marks_value)
        new_count += 1
        
        print(f"✓ Added {subject_input} with {marks_value} marks")
    
    else:
        print("Invalid choice. Please enter 1 or 2")
```

```python
print("\nNew subjects added:", new_count)

updated_avg = round(sum(marks) / len(marks), 2)
print("Updated Average   :", updated_avg)
```

# **#Task 3**

```python
# Data provided in the Task 3

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
#Preocessing + Report table

pass_count = 0
fail_count = 0

avg_list = []
topper_name = ""
topper_avg = 0

print("Name            | Average | Status")
print("----------------------------------------")

for student in class_data:
    name = student[0]
    marks = student[1]
    
    avg = round(sum(marks) / len(marks), 2)
    avg_list.append(avg)
    
    # Pass/Fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1
    
    # Topper logic
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name
    
    # Formatted row
    print(f"{name:<15} | {avg:>7} | {status}")
```

```python
#Summary 

class_avg = round(sum(avg_list) / len(avg_list), 2)

print("\nSummary:")
print("Passed Students :", pass_count)
print("Failed Students :", fail_count)
print("Class Topper    :", topper_name, "-", topper_avg)
print("Class Average   :", class_avg)
```
